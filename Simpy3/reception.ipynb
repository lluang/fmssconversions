```python
"""
Hospital receptionist example.

Covers:

- Waiting for other processes
- Resources: Resource

Scenario:
  A carwash has a limited number of washing machines and defines
  a washing processes that takes some (random) time.

  Car processes arrive at the carwash at a random time. If one washing
  machine is available, they start the washing process and wait for it
  to finish. If not, they wait until they an use one.

"""
import random
from scipy.stats import expon
import simpy

RANDOM_SEED = 42
NUM_RECEPTIONISTS = 2  # Number of machines in the carwash
AVGRECEPTIONTIME = 5      # Average Minutes it takes to serve a customer
T_INTER = 7       # Avverage patient interarrival time
SIM_TIME = 120     # Simulation time in minutes


class Receptionist(object):
    """A Hospital receptionist desk has a limited number of receptionists (``NUM_MACHINES``) to
    help arriving patients.

    Cars have to request one of the machines. When they got one, they
    can start the washing processes and wait for it to finish (which
    takes ``servicetime`` minutes).

    """
    def __init__(self, env, num_receptionists, servicetime):
        self.env = env
        self.receptionist = simpy.Resource(env, num_receptionists)
        self.servicetime = servicetime

    def service(self, patient):
        """The reception processes. It takes an arriving patient and services it."""
        receptiontime = expon.rvs(0, AVGRECEPTIONTIME)
        yield self.env.timeout(receptiontime)
        print("Patient on her way")


def patient(env, name, reception):
    """The patient (each patient has a ``name``) arrives at the receptiondesk
    (``reception``) and waits for a receptionist.

    She is then served and goes on her way.

    """
    print('%s arrives at the reception desk at %.2f.' % (name, env.now))
    with reception.receptionist.request() as request:
        arrivaltime = env.now
        yield request

        print('%s arrives at the receptionist at at %.2f.' % (name, env.now))
        yield env.process(reception.service(name))
        
        leavetime = env.now

        print('%s leaves the receptionist at %.2f.' % (name, leavetime))
        totalwait = leavetime-arrivaltime
        Rstats.waittime.append(totalwait)


def setup(env, num_receptionists, servicetime, t_inter):
    """Create the receptionist desk, a number of initial patients and keep creating patients
    approx. every ``t_inter`` minutes."""
    # Create the carwash
    receptiondesk = Receptionist(env, num_receptionists, servicetime)

    # Create 4 initial patients
    for i in range(4):
        env.process(patient(env, 'Patient %d' % i, receptiondesk))

    # Create more patients while the simulation is running
    while True:
        yield env.timeout(expon.rvs(t_inter))
        i += 1
        env.process(patient(env, 'Patient %d' % i, receptiondesk))

class Rstats:
    waittime = []

# Setup and start the simulation
print('Reception desk')
random.seed(RANDOM_SEED)  # This helps reproducing the results

# Create an environment and start the setup process
env = simpy.Environment()
env.process(setup(env, NUM_RECEPTIONISTS, AVGRECEPTIONTIME, T_INTER))

# Execute!
env.run(until=SIM_TIME)
```
