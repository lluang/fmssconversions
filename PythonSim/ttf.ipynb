```python
import SimPy.Simulation as Sim
import random
```

```python
class Failure(Sim.Process):
    """ Source generates customers randomly """

    def generate(self, component):
        i = 0
        while(component.n>0):
            t = random.randint(1, 6)
            yield Sim.hold, self, t
            j = Job(name="Job%02d" % (i), sim=self.sim)
            self.sim.activate(j, j.visit(component))
            i += 1
        return self.sim.now()
            
class Job(Sim.Process):
    """ Fax arrives and is procossed 

        Processed first by regular staff, then specialized staff if needed
    """

    def visit(self, component):
        arrive = self.sim.now()
        yield Sim.request, self, component
        yield Sim.hold, self, T.repairtime
```

```python
class T:
    repairtime = 2.5
    numcomponents = 2
    maxTime = 100
class TTF(Sim.Simulation):
    def run(self, aseed):
        random.seed(aseed)
        components = Sim.Resource(capacity=T.numcomponents, 
            name="Components", unitName="Component", monitored = True, sim=self) 
        self.initialize()
        f = Failure('Source', sim=self)
        self.ttfmon = Sim.Monitor()
        self.activate(f, f.generate(components))
        self.simulate(until = T.maxTime)
```

```python
ttfsim = TTF()
ttfsim.run(123456)
```
